```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
dataset = fetch_olivetti_faces(shuffle=True, random_state=42)
```

```python
X, y = dataset.images, dataset.target
```

```python
X.shape, y.shape
```

```python
X = X.reshape(400, 4096)
```

```python
X.shape
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.4, random_state=42)
for train_index, test_index in split.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
split2 = StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=42)
for test_index, val_index in split2.split(X_test, y_test):
    X_test, X_val = X_test[test_index], X_test[val_index]
    y_test, y_val = y_test[test_index], y_test[val_index]
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
X_val.shape, y_val.shape
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"n_clusters": range(30, 130)}
```

```python
grid_clf = GridSearchCV(kmeans, param_grid, cv=3, verbose=2)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.99)
X_train_pca = pca.fit_transform(X_train)
X_val_pca = pca.transform(X_val)
X_test_pca = pca.transform(X_test)

pca.n_components_
```

```python
grid_clf.fit(X_train_pca)
```

```python
grid_clf.best_params_
```

```python
from sklearn.metrics import silhouette_score
```

```python
n_clusters = list(range(30, 130))
silhouette_scores = []
interia_scores = []
min_intertia = float('Inf')
max_silhoutte = float('-Inf')
best_k = None
```

```python
for k in n_clusters:
    kmeans.n_clusters = k
    kmeans.fit(X_train_pca)
    current_silhouette_score = silhouette_score(X_train_pca, kmeans.labels_)
    if current_silhouette_score > max_silhoutte:
        max_silhoutte = current_silhouette_score
        best_k = k
    silhouette_scores.append(current_silhouette_score)
```

```python
best_k, max_silhoutte
```

```python
for k in n_clusters:
    kmeans.n_clusters = k
    kmeans.fit(X_train_pca)
    if kmeans.inertia_ < min_intertia:
        min_intertia = kmeans.inertia_
        best_k = k
    interia_scores.append(kmeans.inertia_)
```

```python
best_k, min_intertia
```

```python
import seaborn as sns
```

```python
def plot_faces(faces, labels, n_cols=5):
    n_rows = (len(faces) - 1) // n_cols + 1
    plt.figure(figsize=(n_cols, n_rows * 1.1))
    for index, (face, label) in enumerate(zip(faces, labels)):
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(face.reshape(64, 64), cmap="gray")
        plt.axis("off")
        plt.title(label)
```

```python
import numpy as np
for cluster_id in np.unique(kmeans.labels_):
    print("Cluster", cluster_id)
    in_cluster = kmeans.labels_==cluster_id
    faces = X_train[in_cluster].reshape(-1, 64, 64)
    labels = y_train[in_cluster]
    plot_faces(faces, labels)
```
